```python
import pandas as pd

# Load both CSVs
fred_data = pd.read_csv("fred_data.csv")
```

```python
len(fred_data)
```

```python
fred_reversal = pd.read_csv("fred_reversal.csv", usecols=["result", "description"])

# Rename 'result' column in fred_reversal
fred_reversal = fred_reversal.rename(columns={"result": "result_alt"})

# Merge based on matching 'description' values
merged = fred_data.merge(
    fred_reversal[["description", "result_alt"]],
    on="description",
    how="left"
)

# Save or inspect
merged.to_csv("fred_merged.csv", index=False)
```
