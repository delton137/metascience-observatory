```python
import os
import pandas as pd
#from paperscraper.pdf import save_pdf
from tqdm import tqdm
from fetch_pdf_from_doi import fetch_pdf_from_doi

df = pd.read_csv("replications_database_2025_11_11_180242.csv")
```

```python

doi_column = "original_url"
#save_dir = "/home/dan/Dropbox/AAA_METASCIENCE_OBSERVATORY/PDFs/in_ground_truth_dataset/"
save_dir = "/Users/dan/Library/CloudStorage/Dropbox/AAA_METASCIENCE_OBSERVATORY/PDFs/in_ground_truth_dataset/"

urls = df[doi_column].dropna().astype(str).unique()

missing_dois = []

for url in tqdm(urls, desc="Fetching PDFs"):

    if "doi" in url:
        doi = url.replace("http://doi.org/", "")
        doi = doi.replace("https://doi.org/", "") #for https
        doi = doi.strip(" ")
        doi = doi.strip().lower()
            
        safe_filename = doi.replace("/", "--") + ".pdf"
        pdf_path = os.path.join(save_dir, safe_filename)
    
        if os.path.exists(pdf_path):
            continue

        if os.path.exists(pdf_path):
            continue
        else:
            print(f"------{doi}")
    
            # Try fetching
            result = fetch_pdf_from_doi(doi, pdf_path, verbose=True)
        
            if not result or not os.path.exists(result):
                missing_dois.append(doi)


    else:
        print(f"don't know what to do with {url}")

```
